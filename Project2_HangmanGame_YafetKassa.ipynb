```python
import random

 
name = input("Print name here ")
# This is just a general greeting done by user inputing their name
print("Welcome "  +name+  " to the Hangman Game by Yafet!!!")
print("This is not regular hangman, you will try to decode one of 10 encrypted codes, and you will have ten tries per each code")
print( "GET READY")
 
# These are the codes
words = ['WHDSPQZ', 'XHO', 'TTDBFRT', 'QQJYF', 'ENQD', 'DNPK', 'CNTR', 'EHWHOD', 'TSVMOHOF', 'XNOCFQGTM'] 
 # This will randomly select a word
word = random.choice(words)
 
print("Guess the letters")
 
guesses = ''
 
turns = 10
 
 #This while loop checks if turns are still availible
while turns > 0:
     
    failed = 0
     
    for char in word: 
         
        if char in guesses: 
            print(char)
             
        else: 
            print("_")
             
            failed += 1
             
 
    if failed == 0:
        print("You Win") 
         
        print("Encrypted word is: ", word) 
        print("The decrypted word is", decrypt(word))
        break
     
    guess = input("guess a letter:")
      
    guesses += guess 
     
    if guess not in word:
         
        turns -= 1
          
        print("Incorrect")
         
        print( + turns, 'chances remain')
         
         
        if turns == 0:
            print("GAME OVER!!")
```

```python
def decrypt(ciphertext):
    plaintext = ""
    i=0
    for i in range (0,len(ciphertext)):
        if i%2 == 0:
            plain_ascii = ciphertext[i]
            cipher_ascii = ord(plain_ascii) - 1
            plaintext = plaintext + str(chr(cipher_ascii))

        if i%2 == 1:
            plain_ascii = ciphertext[i]
            cipher_ascii = ord(plain_ascii) + 1
            plaintext = plaintext + str(chr(cipher_ascii))

    return plaintext  
```
